```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Usuário': ['Usuário 01', 'Usuário 02', 'Usuário 03', 'Usuário 04', 'Usuário 05'],
    '1.1': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '1.2': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '1.3': ['Não', 'Sim', 'Não', 'Não', 'Não'],
    '1.4': ['Sim', 'Sim', 'Sim', 'Sim', 'Não'],
    '1.5': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '1.6': ['Sim', 'Não', 'Sim', 'Não', 'Sim'],
    '2.1': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '2.2': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '2.3': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '2.4': ['Sim', 'Sim', 'Não', 'Não', 'Sim'],
    '2.5': ['Sim', 'Sim', 'Sim', 'Não', 'Sim'],
    '2.6': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '3.1': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '3.2': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '3.3': ['Sim', 'Sim', 'Sim', 'Sim', 'Sim'],
    '3.4': ['Sim', 'Sim', 'Sim', 'Não', 'Sim'],
    '3.5': ['Não', 'Não', 'Não', 'Não', 'Não']
}

df = pd.DataFrame(data)
df.replace({'Sim': 1, 'Não': 0}, inplace=True)

task_results_yes = df.sum().drop("Usuário")  
task_results_no = len(df) - task_results_yes  

tasks = ['Tarefa 1', 'Tarefa 2', 'Tarefa 3']
subtasks = [6, 6, 5] 
indices = np.cumsum([0] + subtasks)

fig, axes = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

for i, (start, end) in enumerate(zip(indices[:-1], indices[1:])):
    ax = axes[i]
    bar_width = 0.5
    
    task_yes = task_results_yes[start:end]
    task_no = task_results_no[start:end]
    
    if i == 0:
        ax.bar(range(len(task_yes)), task_yes, width=bar_width, color='#ac89cf', label="Sim")
        ax.bar(range(len(task_no)), task_no, width=bar_width, bottom=task_yes, color='#5c20a6', label="Não")
    else:
        ax.bar(range(len(task_yes)), task_yes, width=bar_width, color='#ac89cf')
        ax.bar(range(len(task_no)), task_no, width=bar_width, bottom=task_yes, color='#5c20a6')
    
    ax.set_xlabel("Subtarefas")
    ax.set_xticks(range(len(task_yes)))
    ax.set_xticklabels(task_yes.index, rotation=45, ha="right")
    ax.set_title(tasks[i])

axes[0].set_ylabel("Número de Usuários")
fig.tight_layout(rect=[0, 0.1, 1, 0.95])  
fig.legend(title="Conclusão", loc="lower center", ncol=2, bbox_to_anchor=(0.5, -0.05))
plt.show()
```
