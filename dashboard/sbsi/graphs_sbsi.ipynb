```python
import matplotlib.pyplot as plt
import pandas as pd

def time_to_seconds(time_str):
    minutes, seconds = map(int, time_str.split(":"))
    return minutes * 60 + seconds

colors = {0: '#8dd3c7', 1: '#0d8007', 2: '#984397', 3: '#fb8072', 4: '#80b1d3', 5: '#fdb462'}

data1 = {
    "Participant 01": ["01:30", "03:27", "01:00", "03:48", "01:30", "00:30"],
    "Participant 02": ["01:30", "00:30", "02:35", "03:30", "04:40", "00:40"],
    "Participant 03": ["01:30", "00:47", "03:20", "02:35", "03:50", "00:30"],
    "Participant 04": ["01:30", "00:35", "01:35", "03:15", "02:40", "00:30"],
    "Participant 05": ["01:30", "01:25", "02:20", "02:30", "02:37", "00:30"]
}
data2 = {
    "Participant 01": ["01:00", "03:25", "01:15", "04:10", "01:15", "00:50"],
    "Participant 02": ["00:30", "04:28", "00:20", "04:50", "01:00", "01:23"],
    "Participant 03": ["00:37", "03:26", "00:30", "03:30", "02:15", "01:55"],
    "Participant 04": ["00:25", "04:50", "00:40", "02:55", "01:35", "02:12"],
    "Participant 05": ["00:30", "03:40", "00:20", "03:35", "02:37", "00:45"]
}
data3 = {
    "Participant 01": ["01:15", "01:00", "02:50", "03:20", "04:35"],
    "Participant 02": ["00:30", "01:25", "01:00", "04:27", "05:35"],
    "Participant 03": ["00:30", "01:57", "01:00", "03:50", "02:54"],
    "Participant 04": ["00:25", "02:15", "01:00", "03:34", "03:40"],
    "Participant 05": ["00:30", "02:35", "01:00", "04:50", "04:55"]
}

dfs = [
    pd.DataFrame(data1, index=["Task 1.1", "Task 1.2", "Task 1.3", "Task 1.4", "Task 1.5", "Task 1.6"]),
    pd.DataFrame(data2, index=["Task 2.1", "Task 2.2", "Task 2.3", "Task 2.4", "Task 2.5", "Task 2.6"]),
    pd.DataFrame(data3, index=["Task 3.1", "Task 3.2", "Task 3.3", "Task 3.4", "Task 3.5"])
]

titles = [
    "Task 1 Realization Time by Participants",
    "Task 2 Realization Time by Participants",
    "Task 3 Realization Time by Participants"
]

fig, axes = plt.subplots(nrows=3, figsize=(10, 14))

for ax, df, title in zip(axes, dfs, titles):
    df_sec = df.applymap(time_to_seconds)
    for i, user in enumerate(df_sec.columns, start=1):
        ax.scatter(df_sec[user], df_sec.index, color=colors[i], label=user, s=100, zorder=2)
    ax.set_title(title)
    ax.set_xlabel("Time (seconds)")
    ax.grid(axis='y', linestyle='--', alpha=0.7)

handles, labels = axes[-1].get_legend_handles_labels()
fig.legend(
    handles, labels,
    title="Participants",
    loc="lower center",
    bbox_to_anchor=(0.5, -0.03),
    ncol=3,
    frameon=True,
    handletextpad=0.5,
    columnspacing=0.5,
    borderaxespad=0.6
).get_frame().set_facecolor('#f1f1f1')

fig.subplots_adjust(bottom=0.1, top=1, hspace=0.3)
plt.savefig(r"C:\Users\annab\OneDrive\Documentos\GitHub\Colabs\dashboard\img/all_tasks_time_stacked.png", format="png", bbox_inches='tight', dpi=600)
plt.show()
```
