```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from sklearn.metrics import confusion_matrix
```

# DT

```python
def create_conf_matrix():
    true_positive_0 = 7805  
    false_positive_0 = 3592  
    false_positive_1 = 3925  
    true_positive_1 = 7472 
    
    conf_matrix = np.array([[true_positive_0, false_positive_0],
                            [false_positive_1, true_positive_1]])
    
    colors = ['#fff', '#b9cad8', '#56728c', '#021c40']
    cmap = LinearSegmentedColormap.from_list("custom_blue_gradient", colors)
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap=cmap,  # 'd' para valores inteiros
                xticklabels=["Classe 0", "Classe 1"],
                yticklabels=["Classe 0", "Classe 1"],
                cbar=True)
    
    plt.xlabel('Labels Previstos')
    plt.ylabel('Labels Verdadeiros')
    #plt.title('Matriz de Confus√£o com Gradiente Personalizado')
    plt.show()

create_conf_matrix()
```
