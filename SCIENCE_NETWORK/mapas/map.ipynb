# Mapa

```python
import folium
```

```python
df = df.dropna(subset=['latitude', 'longitude'])
df['premature'] = df['GESTACAO'].apply(lambda x: 1 if x >= 4 else 0)

mapa = folium.Map(location=[-8.0476, -34.8813], zoom_start=7)

def adicionar_ponto(row):
    cor = 'red' if row['premature'] == 1 else 'blue'  
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        color=cor,
        fill=True,
        fill_color=cor, 
        fill_opacity=0.7,
        popup=f"Município: {row['CODMUNNASC']}\nPrematuro: {'Sim' if row['premature'] == 1 else 'Não'}"
    ).add_to(mapa)

df.apply(adicionar_ponto, axis=1)
mapa.save('mapa_nascimentos_prematuros.html')
mapa
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

df_prematuros = df[df['GESTACAO'] >= 4]
if 'longitude' not in df_prematuros.columns or 'latitude' not in df_prematuros.columns or 'SEMAGESTAC_por_100k' not in df_prematuros.columns:
    raise ValueError("O dataset deve conter as colunas 'longitude', 'latitude' e 'SEMAGESTAC_por_100k'.")

geometry = [Point(xy) for xy in zip(df_prematuros['longitude'], df_prematuros['latitude'])]
gdf_prematuros = gpd.GeoDataFrame(df_prematuros, geometry=geometry)

gdf_prematuros.set_crs(epsg=4326, inplace=True)

shapefile_path = '/content/drive/MyDrive/Mestrado/Ciencia das redes/PE_Municipios_2022/PE_Municipios_2022.shp'
municipios_gdf = gpd.read_file(shapefile_path)
fig, ax = plt.subplots(figsize=(12, 12))

municipios_gdf.plot(ax=ax, color='whitesmoke', edgecolor='black', alpha=0.7)
bubble_size = gdf_prematuros['SEMAGESTAC_por_100k'] / 2

scatter = ax.scatter(
    gdf_prematuros['longitude'],
    gdf_prematuros['latitude'],
    s=bubble_size,
    c=gdf_prematuros['SEMAGESTAC_por_100k'],
    cmap='Reds',
    alpha=0.6,
    edgecolor='k',
    linewidth=0.5
)

cbar = plt.colorbar(scatter, ax=ax, shrink=0.7)  # Reduzir o tamanho da barra lateral usando 'shrink'
cbar.set_label('Casos de Nascimentos Prematuros por 100k', fontsize=12)

plt.title('Distribuição Proporcional de Nascimentos Prematuros', fontsize=15)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)

plt.show()
```
